```python
df_NASC_sel = df_NASC_sel.dropna(subset=["longitude", "latitude", "NASC"])
df_NASC_sel
```

```python
import geoviews as gv
import geoviews.tile_sources as gvts
```

```python
def scale_sizes(values, min_value, max_value, min_size=25, max_size=250):

    # Censor values if needed
    sizes = values.copy()
    sizes.loc[sizes < min_value] = min_value
    sizes.loc[sizes > max_value] = max_value

    return ((sizes - min_value) / (max_value - min_value)) * (max_size - min_size) + min_size
```

```python
df_NASC_sel["size"] = scale_sizes(
    df_NASC_sel["NASC"],
    min_value=1e2,
    max_value=1e4,
    min_size=5,
    max_size=20,
)
```

```python
df_NASC_sel
```

```python
# Remove 'channel' column if present
if 'channel' in df_NASC_sel.columns:
    df_NASC_sel = df_NASC_sel.drop(columns=['channel'])

# Format 'ping_time' as ISO8601 string if present
if 'ping_time' in df_NASC_sel.columns:
    df_NASC_sel['ping_time'] = pd.to_datetime(df_NASC_sel['ping_time'], format='ISO8601', errors='coerce').dt.strftime('%Y-%m-%dT%H:%M:%S')

# Only display NASC, latitude, and longitude in hover tooltip
points = gv.Points(df_NASC_sel, ['longitude', 'latitude'], ['NASC', "size", "ping_time", "filename"]).opts(
    color='NASC',
    cmap='Reds',
    clim=(1e2, 1e4),
    colorbar=True,
    size='size',
    alpha=0.5,
    width=900,
    height=800,
    tools=['hover'],
    xlabel='Longitude',
    ylabel='Latitude',
    title='NASC Geo Scatter',
    hover_tooltips=[
        ('Longitude', '@longitude'),
        ('Latitude', '@latitude'),
        ('NASC', '@NASC'),
        ('Ping Time', '@ping_time'),
        ('Filename', '@filename'),
    ]
)

track = gv.Path([df_NASC[['longitude', 'latitude']].values]).opts(
    color='gray',
    line_width=1
)

# Select basemap
# tile = gvts.EsriNatGeo
tile = gvts.EsriOceanBase

# Overlay everything
overlay = tile * points * track
overlay
```

```python
import panel as pn
```

```python
# Get all WMTS tile source names
tile_options = [name for name, obj in gvts.__dict__.items() if isinstance(obj, gvts.WMTS)]

tile_selector = pn.widgets.Select(
    name="Basemap tile source",
    options=tile_options,
    value="EsriNatGeo"
)
```

```python
@pn.depends(tile_selector)
def plot_track_bio_var(
    map_tile: gv.element.geo.WMTS,
) -> None:
    tile = getattr(gvts, map_tile)
    
    overlay = tile * points * track
    return overlay
```

```python
server2 = pn.serve(
    pn.Column(
        "# Track Map with Bio-variables",
        tile_selector,
        pn.layout.Divider(),
        pn.panel(plot_track_bio_var),
    ),
    port=5010, show=True
)
```

```python
from pathlib import Path
import pandas as pd

import holoviews as hv
import panel as pn
```

```python
path_vm_local = Path("/media/volume/shimada_202506_volume/integration")
```

```python
df_len_cnt_all = pd.read_csv(path_vm_local / "length_count_all.csv", index_col=0)
df_len_cnt_all
```

```python
df_len_cnt_all.hist(column="length", by=["sex","stratum"], bins=30, figsize=(12, 5))
```

```python
import holoviews as hv
import panel as pn
import numpy as np
from bokeh.palettes import Category10
hv.extension('bokeh')
pn.extension()

sex_options = df_len_cnt_all['sex'].unique().tolist()

# Sort stratum numerically
stratum_options = sorted(df_len_cnt_all['stratum'].unique(), key=lambda x: int(x))

# Assign colors for each stratum
palette = Category10[10]  # Up to 10 colors
stratum_colors = {stratum: palette[i % len(palette)] for i, stratum in enumerate(stratum_options)}

def plot_histograms():
    plots = []
    for sex in sex_options:
        overlay = None
        for i, stratum in enumerate(stratum_options):
            df_sub = df_len_cnt_all[(df_len_cnt_all['sex'] == sex) & (df_len_cnt_all['stratum'] == stratum)]
            if not df_sub.empty:
                hist = hv.Histogram(np.histogram(df_sub['length'], bins=np.arange(0, 101, 5)), label=f"Stratum: {stratum}").opts(
                    color=stratum_colors[stratum],
                    alpha=0.6,  # default transparency
                    muted_alpha=0.1,  # transparency when muted via legend
                    height=250, width=500,
                )
                overlay = hist if overlay is None else overlay * hist
        if overlay:
            overlay = overlay.opts(title=f"Sex: {sex}", legend_position='right')
            plots.append(overlay)
    if plots:
        return hv.Layout(plots).cols(1)
    else:
        return hv.Text(0, 0, "No data for available strata.")

layout = pn.Column(
    pn.pane.Markdown("# Length histograms"),
    pn.Row(
        plot_histograms,
    )
)

# layout.servable()
server2 = pn.serve(layout, port=5008, show=True)
```

```python
import re
```

```python
filenames = ["IWCPS-D20250613-T115151.raw", "IWCPS-D20250613-T115748.raw"]
```

```python
# Extract datetime objects from filenames
def extract_datetime_from_filename(filename):
    match = re.search(r'D(\d{8})-T(\d{6})', filename)
    if match:
        date_str = match.group(1)
        time_str = match.group(2)
        return datetime.datetime.strptime(f"{date_str} {time_str}", "%Y%m%d %H%M%S").replace(tzinfo=datetime.timezone.utc)
    return None
```

```python
[extract_datetime_from_filename(filename) for filename in filenames]
```

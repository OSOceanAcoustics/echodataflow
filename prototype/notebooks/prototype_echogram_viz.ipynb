```python
from pathlib import Path
import xarray as xr
import numpy as np
```

```python
ds_MVBS1 = xr.Dataset(
    data_vars={
        "Sv": (["channel", "ping_time", "echo_range"], -70*np.ones((5, 1693, 760))),
    },
    coords={
        "channel": [
            "WBT 400141-15 ES18_ES",
            "WBT 400143-15 ES38B_ES",
            "WBT 400142-15 ES70-7C_ES",
            "WBT 400140-15 ES120-7C_ES",
            "WBT 400145-15 ES200-7C_ES",
        ],
        "echo_range": np.arange(760),
        "ping_time": np.arange(1693),
    },
)
ds_MVBS1["Sv"] = ds_MVBS1["Sv"].assign_attrs(
    actual_range=(
        float(ds_MVBS1["Sv"].min().compute()),
        float(ds_MVBS1["Sv"].max().compute())
    )
)
```

```python
ds_MVBS2 = xr.Dataset(
    data_vars={
        "Sv": (["channel", "ping_time", "echo_range"], -50*np.ones((5, 5387, 760))),
    },
    coords={
        "channel": [
            "WBT 400141-15 ES18_ES",
            "WBT 400143-15 ES38B_ES",
            "WBT 400142-15 ES70-7C_ES",
            "WBT 400140-15 ES120-7C_ES",
            "WBT 400145-15 ES200-7C_ES",
        ],
        "echo_range": np.arange(760),
        "ping_time": np.arange(5387),
    },
)
ds_MVBS2["Sv"] = ds_MVBS2["Sv"].assign_attrs(
    actual_range=(
        float(ds_MVBS2["Sv"].min().compute()),
        float(ds_MVBS2["Sv"].max().compute())
    )
)
```

```python
ds_MVBS1.chunk(
    {"channel": -1, "ping_time": -1, "echo_range": -1}
).to_zarr(
    "/Users/wujung/Downloads/latest_MVBS.zarr",
    mode="w",
    consolidated=True,
)
```
